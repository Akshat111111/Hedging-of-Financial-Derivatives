#### Let's make a volatility function.

```python
def get_daily_volatility(df):

    # Calculate daily returns
    daily_returns = df['Adj Close'].pct_change().dropna()

    # Calculate daily volatility (standard deviation of daily returns)
    daily_volatility = daily_returns.std()

    return daily_volatility

def get_monthly_volatility(df):

    # Calculate monthly returns
    monthly_returns = df['Adj Close'].resample('M').ffill().pct_change().dropna()

    # Calculate monthly volatility (standard deviation of monthly returns)
    monthly_volatility = monthly_returns.std()

    return monthly_volatility
```

```python
x = get_daily_volatility(ni), get_daily_volatility(sp500), get_daily_volatility(nasdaq), get_daily_volatility(nse)
x
```

```python
y = get_daily_volatility(btc), get_daily_volatility(eth), get_daily_volatility(bnb)
y
```

```python
z = np.append(np.array(x), np.array(y))
z
```

### Plotting

```python
plt.scatter(['NIKKEI','S&P500','Nasdaq','NIFTY 50','Bitcoin','Ethereum','BNB'],z,s=10,alpha=0.75,c='g')
plt.ylabel('Volatiliy')
```

#### Thus, cryptos have a higher volatility than stocks.

# Additional visualizations

```python
# Helper function to plot moving averages
def plot_moving_averages(data, title, short_window=50, long_window=200):
    plt.figure(figsize=(12, 6))
    plt.plot(data['Adj Close'], label='Actual Prices', alpha=0.5)
    plt.plot(data['Adj Close'].rolling(window=short_window).mean(), label=f'{short_window}-Day Moving Average')
    plt.plot(data['Adj Close'].rolling(window=long_window).mean(), label=f'{long_window}-Day Moving Average')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Helper function to plot Bollinger Bands
def plot_bollinger_bands(data, title, window=20):
    rolling_mean = data['Adj Close'].rolling(window).mean()
    rolling_std = data['Adj Close'].rolling(window).std()

    plt.figure(figsize=(12, 6))
    plt.plot(data['Adj Close'], label='Actual Prices', alpha=0.5)
    plt.plot(rolling_mean, label='Rolling Mean', color='orange')
    plt.fill_between(data.index, rolling_mean - 2*rolling_std, rolling_mean + 2*rolling_std, color='orange', alpha=0.2, label='Bollinger Bands')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Comparative close prices
def plot_comparative_close(data_dict, title):
    plt.figure(figsize=(12, 6))
    for label, data in data_dict.items():
        plt.plot(data['Adj Close'], label=label)
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Volatility plot
def get_volatility(df, period='D'):
    if period == 'D':
        returns = df['Adj Close'].pct_change().dropna()
    elif period == 'M':
        returns = df['Adj Close'].resample('M').ffill().pct_change().dropna()
    return returns.std()
```

```python
# Plot moving averages and Bollinger Bands for S&P 500 and Bitcoin
plot_moving_averages(sp500, 'S&P 500 Moving Averages')
plot_bollinger_bands(sp500, 'S&P 500 Bollinger Bands')
plot_moving_averages(btc, 'Bitcoin Moving Averages')
plot_bollinger_bands(btc, 'Bitcoin Bollinger Bands')
```

```python
# Comparative close prices for indices
plot_comparative_close({'S&P 500': sp500, 'NIFTY 50': nse, 'NIKKEI': ni, 'Nasdaq': nasdaq}, 'Comparative Close Prices of Indices')
```

```python
# Comparative close prices for cryptocurrencies
plot_comparative_close({'Bitcoin': btc, 'Ethereum': eth, 'BNB': bnb}, 'Comparative Close Prices of Cryptocurrencies')
```

```python
# Volatility comparison
stocks_volatility = [get_volatility(ni), get_volatility(sp500), get_volatility(nasdaq), get_volatility(nse)]
cryptos_volatility = [get_volatility(btc), get_volatility(eth), get_volatility(bnb)]
combined_volatility = np.append(stocks_volatility, cryptos_volatility)

labels = ['NIKKEI', 'S&P 500', 'Nasdaq', 'NIFTY 50', 'Bitcoin', 'Ethereum', 'BNB']
plt.figure(figsize=(12, 6))
plt.bar(labels, combined_volatility, color='skyblue')
plt.title('Volatility Comparison')
plt.ylabel('Volatility')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
