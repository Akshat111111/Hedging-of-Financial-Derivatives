## Hedging Currencies

### Hedging currency involves strategies such as forward contracts, options, or simply balancing your exposure by using financial instruments available in the forex market.<br>Here, I'll outline a basic approach to hedge currency risk using forward contracts.

```python
# Imports
import matplotlib.pyplot as plt
import yfinance as yf
import seaborn as sns
import pandas as pd
```

### Hedging Euro

```python
# Loading the data
df1 = yf.download('USDEUR=X', start='2024-01-01', end='2024-05-14').drop('Volume',axis=1)
df1
```

```python
# Plotting currency
plt.plot(df1['Adj Close'])
plt.xticks(rotation=45)
plt.title('USD vs EUR')
plt.xlabel('Date')
plt.ylabel('USD(in EUR)')

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plotting volatility of the currency
plt.plot(df1['Adj Close'].pct_change())
plt.xticks(rotation=45)
plt.title('Volatility of EUR')
plt.xlabel('Date')
plt.ylabel('Change')

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
base_currency = 'USD'
target_currency = 'EUR'
current_rate = 0.92
exposure_amount = 100000

# Function to calculate the amount needed for hedging
def calculate_hedge_amount(exposure_amount, current_rate):
    return exposure_amount / current_rate

# Get the current exchange rate
print(f"Current exchange rate from {base_currency} to {target_currency}: {current_rate}")

# Calculate hedge amount#
hedge_amount = calculate_hedge_amount(exposure_amount, current_rate)
print(f"Amount to hedge: {hedge_amount} {base_currency}")

# Simulating a forward contract rate
def simulate_forward_rate(current_rate, base_currency, target_currency, period_days):
    base_rate = 0.01  # Example base currency interest rate
    target_rate = 0.02  # Example target currency interest rate
    forward_rate = current_rate * (1 + (target_rate - base_rate) * (period_days / 365))
    return forward_rate

# Define the period for the forward contract
period_days = 90 # days

# Calculate the forward rate
forward_rate = simulate_forward_rate(current_rate, base_currency, target_currency, period_days)
print(f"Simulated forward rate for {period_days} days: {forward_rate}")

# Calculate future value of exposure using forward rate
future_value = hedge_amount * forward_rate
print(f"Future value of the exposure using forward rate: {future_value} {target_currency}")
```

```python
# Function to compare historical exchange rates
def compare_exchange_rates(base_currency, target_currencies, start_date, end_date):
    plt.figure(figsize=(15, 8))
    for currency in target_currencies:
        df = yf.download(f'{currency}{base_currency}=X', start=start_date, end=end_date)['Adj Close']
        plt.plot(df, label=f'{currency}/{base_currency}')
    plt.title('Historical Exchange Rate Comparison')
    plt.xlabel('Date')
    plt.ylabel('Exchange Rate')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
# Function to perform correlation analysis
def correlation_analysis(base_currency, target_currencies, start_date, end_date):
    plt.figure(figsize=(12, 8))
    data = {}
    for currency in target_currencies:
        df = yf.download(f'{currency}{base_currency}=X', start=start_date, end=end_date)['Adj Close']
        data[currency] = df
    df_corr = pd.DataFrame(data)
    sns.heatmap(df_corr.corr(), annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Correlation Analysis')
    plt.xlabel('Currencies')
    plt.ylabel('Currencies')
    plt.tight_layout()
    plt.show()
```

```python
# Currency pairs and date range for analysis
base_currency = 'USD'
target_currencies = ['EUR', 'INR', 'JPY', 'RUB', 'CNY']
start_date = '2024-01-01'
end_date = '2024-05-14'
```

```python
# Perform correlation analysis
correlation_analysis(base_currency, target_currencies, start_date, end_date)
```

```python
# Compare historical exchange rates
compare_exchange_rates(base_currency, target_currencies, start_date, end_date)
```
