```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("rent_ads_rightmove_extended.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
# Set the style for seaborn plots
sns.set(style="whitegrid")
```

```python
# Visualization 1: Distribution of Rental Prices
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='rent', bins=30, color='orange')
plt.title('Distribution of Rental Prices')
plt.xlabel('Rental Price ()')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 2: Rental Prices vs. Property Type
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='PROPERTY TYPE', y='rent', color='orange')
plt.title('Rental Prices by Property Type')
plt.xlabel('Property Type')
plt.ylabel('Rental Price ()')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualization 3: Average Distance to Nearest Station vs. Rental Prices
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='avg_distance_to_nearest_station', y='rent', color='orange')
plt.title('Average Distance to Nearest Station vs. Rental Prices')
plt.xlabel('Average Distance to Nearest Station (miles)')
plt.ylabel('Rental Price ()')
plt.show()
```

```python
# Visualization 4: Number of Bedrooms Distribution
plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='BEDROOMS', color='orange')
plt.title('Distribution of Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 5: Rental Prices Distribution by Furnish Type
plt.figure(figsize=(10, 6))
sns.violinplot(data=data, x='Furnish Type', y='rent', color='orange')
plt.title('Rental Prices Distribution by Furnish Type')
plt.xlabel('Furnish Type')
plt.ylabel('Rental Price ()')
plt.xticks(rotation=90)
plt.show()
```

```python
# Visualization 6: Relationship between Rental Price and Deposit Amount
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='rent', y='deposit', color='orange')
plt.title('Relationship between Rental Price and Deposit Amount')
plt.xlabel('Rental Price ()')
plt.ylabel('Deposit Amount ()')
plt.show()
```

```python
# Visualization 7: Rental Prices by Council Tax
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Coucil Tax', y='rent', color='orange')
plt.title('Rental Prices by Council Tax')
plt.xlabel('Council Tax')
plt.ylabel('Rental Price ()')
plt.show()
```

```python
# Visualization 8: Rental Prices by Let Type
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Let type', y='rent', color='orange')
plt.title('Rental Prices by Let Type')
plt.xlabel('Let Type')
plt.ylabel('Rental Price ()')
plt.show()
```

```python
# Visualization 9: Rental Prices Distribution by Number of Bathrooms
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='BATHROOMS', y='rent', color='orange')
plt.title('Rental Prices Distribution by Number of Bathrooms')
plt.xlabel('Number of Bathrooms')
plt.ylabel('Rental Price ()')
plt.show()
```

```python
# Visualization 10: Average Distance to Nearest Station Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='avg_distance_to_nearest_station', bins=30, color='orange')
plt.title('Distribution of Average Distance to Nearest Station')
plt.xlabel('Average Distance to Nearest Station (miles)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 11: Relationship between Rental Price and Nearest Station Count
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='nearest_station_count', y='rent', color='orange')
plt.title('Relationship between Rental Price and Nearest Station Count')
plt.xlabel('Nearest Station Count')
plt.ylabel('Rental Price ()')
plt.show()
```

```python
# Visualization 12: Rental Prices Distribution by Number of Bedrooms
plt.figure(figsize=(10, 6))
sns.violinplot(data=data, x='BEDROOMS', y='rent', color='orange')
plt.title('Rental Prices Distribution by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Rental Price ()')
plt.show()
```
