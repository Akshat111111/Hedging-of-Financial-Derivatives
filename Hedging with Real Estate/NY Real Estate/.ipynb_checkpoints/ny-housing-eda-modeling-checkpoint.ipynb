# 3. Modeling

## 3.1 Dropping unnecessary columns

```python
df = df.drop(['type', 'place', 'sublocality'], axis=1)
df.head()
```

```python
X = df.drop(['price'], axis=1)
y = df['price']
```

```python
X.head()
```

```python
y.head()
```

## 3.2 Encoding categorical variables

```python
X_encoded = pd.get_dummies(X, columns=['bath',
                                       'beds'])

X_encoded.head()
```

## 3.3 Splitting data into train and test

```python
X_train, X_test, y_train, y_test = train_test_split(X_encoded, 
                                                    y, 
                                                    random_state=42,)
```

```python
clf_xgb_v1 = xgb.XGBRegressor(objective="reg:squarederror",
                            # missing=None,
                            seed=42)

clf_xgb_v1.fit(X_train,
            y_train,
            verbose=True,
            early_stopping_rounds=20,
            eval_metric='rmse',
            eval_set=[(X_test, y_test)])
```

# 4. Predictions

```python
y_preds = clf_xgb_v1.predict(X_test)
```

```python
fig = plt.figure(figsize=(20, 10))
sns.regplot(data=df, x=y_test, y=y_preds, x_jitter=.15, line_kws=dict(color="r"),)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.ticklabel_format(style = 'plain')
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, y_preds)
```

```python
df = pd.read_csv('/kaggle/input/new-york-housing-market/NY-House-Dataset.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
dic = {}
for col in df.columns:
    dic[col] = col.lower()
dic['PROPERTYSQFT'] = 'area'
    
df.rename(columns = dic, inplace = True)

df.head()
```

```python
df['street_name'].value_counts()
```

```python
df['long_name'].value_counts()
```

```python
df[['street_name', 'long_name']].value_counts()
```

```python
cols_to_drop = ['address', 'address', 'state', 'main_address', 'administrative_area_level_2',
                'formatted_address', 'long_name', 'brokertitle']
df.drop(cols_to_drop, axis = 1, inplace = True)
df
```

```python
X = df.drop('price', axis = 1)
y = df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=77)
```

```python
tmp = X_train[['locality', 'sublocality', 'street_name']].value_counts()
print(tmp)
print(tmp[tmp<30])
```

```python
X_train['loc_tuple'] = list(zip(X_train['locality'],
                                X_train['sublocality'],
                                X_train['street_name']))
X_train.drop(['locality', 'sublocality', 'street_name'], axis = 1, inplace = True)

threshold = 30
frequencies = X_train['loc_tuple'].value_counts()
X_train['OH_Drop'] = X_train['loc_tuple'].apply(lambda x : 1 if frequencies[x] < threshold else 0)

X_train['encoded_loc_tuple'] = X_train.apply(lambda x : 'OH_Drop' if x['OH_Drop'] == 1 
                                                           else x['loc_tuple'], axis=1)



X_test['loc_tuple'] = list(zip(X_test['locality'],
                                X_test['sublocality'],
                                X_test['street_name']))
X_test.drop(['locality', 'sublocality', 'street_name'], axis = 1, inplace = True)

frequencies = X_test['loc_tuple'].value_counts()
X_test['OH_Drop'] = X_test['loc_tuple'].apply(lambda x : 1 if frequencies[x] < threshold else 0)

X_test['encoded_loc_tuple'] = X_test.apply(lambda x : 'OH_Drop' if x['OH_Drop'] == 1 
                                                           else x['loc_tuple'], axis=1)

X_train
```

```python
X_train.encoded_loc_tuple.value_counts()
```

```python
X_train.drop('loc_tuple', axis = 1, inplace = True)
X_test.drop('loc_tuple', axis = 1, inplace = True)
X_train
```

```python
from sklearn.preprocessing import OneHotEncoder

loc_OH_encoder = OneHotEncoder(sparse = False, handle_unknown = 'ignore')

X_train['encoded_loc_tuple'] = X_train['encoded_loc_tuple'].astype(str)
X_test['encoded_loc_tuple'] = X_test['encoded_loc_tuple'].astype(str)

OH_loc_train = pd.DataFrame(loc_OH_encoder.fit_transform(X_train[['encoded_loc_tuple']]),
                            columns = [name for name in
                                    loc_OH_encoder.get_feature_names_out(['encoded_loc_tuple'])])
OH_loc_test = pd.DataFrame(loc_OH_encoder.transform(X_test[['encoded_loc_tuple']]),
                            columns = [name for name in
                                    loc_OH_encoder.get_feature_names_out(['encoded_loc_tuple'])])

OH_loc_train.index = X_train.index
OH_loc_test.index = X_test.index

X_train = pd.concat([X_train, OH_loc_train], axis = 1)
X_train.drop('encoded_loc_tuple', axis = 1, inplace = True)
X_test = pd.concat([X_test, OH_loc_test], axis = 1)
X_test.drop('encoded_loc_tuple', axis = 1, inplace = True)

X_train
```

```python
X_train['type'].value_counts()
```

```python
type_encoder = OneHotEncoder(sparse = False, handle_unknown = 'ignore')

X_train['type'] = X_train['type'].astype(str)
X_test['type'] = X_test['type'].astype(str)

OH_type_train = pd.DataFrame(type_encoder.fit_transform(X_train[['type']]),
                            columns = [name for name in
                                        type_encoder.get_feature_names_out(['type'])])
OH_type_test = pd.DataFrame(type_encoder.transform(X_test[['type']]),
                            columns = [name for name in
                                        type_encoder.get_feature_names_out(['type'])])

OH_type_train.index = X_train.index
OH_type_test.index = X_test.index

X_train = pd.concat([X_train, OH_type_train], axis = 1)
X_train.drop('type', axis = 1, inplace = True)
X_test = pd.concat([X_test, OH_type_test], axis = 1)
X_test.drop('type', axis = 1, inplace = True)

X_train
```

```python
X_train['bath'] = np.ceil(X_train['bath'])
X_test['bath'] = np.ceil(X_test['bath'])
X_train
```

```python
my_model = xgb.XGBRegressor(objective="reg:squarederror",
                            # missing=None,
                            seed=77)

my_model.fit(X_train,
            y_train,
            verbose=True,
            early_stopping_rounds=20,
            eval_metric='rmse',
            eval_set=[(X_test, y_test)])
```

```python
y_preds = my_model.predict(X_test)
```
