```python
import pandas as pd
df = pd.read_csv(filepath_or_buffer='/kaggle/input/grab-holdings-stock-data-all-time/GRAB.csv', parse_dates=['Date'])
df['year'] = df['Date'].dt.year
df.head()
```

```python
from plotly.express import line
line(data_frame=df, x='Date', y=['Open', 'High', 'Low', 'Close', 'Adj Close'], log_y=True)
```

We only have 500 or so quotes so our curve looks pretty smooth.

```python
line(data_frame=df.set_index('Date').resample('W').mean().reset_index(), x='Date', y=['Open', 'High', 'Low', 'Close', 'Adj Close'], log_y=True)
```

The weekly means smooth out the curve and maybe it is a little easier to see how the high and low bound the other values.

```python
line(data_frame=df.set_index('Date').resample('M').mean().reset_index(), x='Date', y=['Open', 'High', 'Low', 'Close', 'Adj Close'], log_y=True)
```

```python
from plotly.express import scatter
scatter(data_frame=df, x='Adj Close', y='Volume', log_x=True, log_y=True, color='year', trendline='lowess')
```

Prices have essentially two regimes, high and low, that are clustered in time; but price and volume are not correlated.
