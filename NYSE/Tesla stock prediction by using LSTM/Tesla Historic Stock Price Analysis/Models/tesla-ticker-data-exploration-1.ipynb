```python
# Lets convert the Date column to datetime format
df['Date']=pd.to_datetime(df['Date'])
# Lets find the starting and ending date of the dataset
starting_date = df['Date'].min()
ending_date   = df['Date'].max()
# Lets check the Format of dates
starting_date_formatted = starting_date.strftime("%Y-%m-%d")
ending_date_formatted = ending_date.strftime("%Y-%m-%d")
# print the starting and ending date
print("Starting date is:", starting_date_formatted)
print("Ending date is:", ending_date_formatted)
```

## Tesla Price By Line Plot

```python
# Lets create a trace for the opening stock market price of Tesla
trace=go.Scatter(x=df['Date'], y=df['Open'], mode='lines', name='Opening price')
# Lets prepare the layout
layout = go.Layout(title='Tesla Stock Market Opening Price', xaxis=dict(title='Date'), yaxis=dict(title='Opening Price'))
fig=go.Figure(data=[trace], layout=layout)
iplot(fig)
```

```python
# Lets create a trace for the closing stock market price of Tesla
trace=go.Scatter(x=df['Date'], y=df['Close'], mode='lines', name='Closing price')
# Lets prepare the layout
layout = go.Layout(title='Tesla Stock Market Closing Price', xaxis=dict(title='Date'), yaxis=dict(title='Closing Price'))
fig=go.Figure(data=[trace], layout=layout)
iplot(fig)
```

## CandleStick Plot

```python
trace=go.Candlestick(x=df['Date'],
                     open=df['Open'],
                     close=df['Close'],
                     high=df['High'],
                     low=df['Low'],
                     name='Tesla Stock')
layout=go.Layout(title='Tesla Stock Market price Candlestick chart',
                 xaxis=dict(title='Date'),
                 yaxis=dict(title='Tesla Stock Price'))
fig=go.Figure(data=[trace], layout=layout)
iplot(fig)
```

## BarChart Plot

```python
# Lets groupby the year and open column to determing the yearly closing price of Tesla market
yearly_data=df.groupby("Year")['Open'].mean().reset_index()
trace=go.Bar(x=yearly_data['Year'], y=yearly_data['Open'],name="Yearly Opening Price")
layout=go.Layout(title='Tesla yearly Opening Price',
                 xaxis=dict(title='Year'),
                 yaxis=dict(title='Opening Price'))
fig=go.Figure(data=[trace], layout=layout)
iplot(fig)
```

```python
# Lets groupby the year and Close column to determing the yearly closing price of Tesla market
yearly_data=df.groupby("Year")['Close'].mean().reset_index()
trace=go.Bar(x=yearly_data['Year'], y=yearly_data['Close'],name="Yearly Closing Price")
layout=go.Layout(title='Tesla yearly Closing Price',
                 xaxis=dict(title='Year'),
                 yaxis=dict(title='Closing Price'))
fig=go.Figure(data=[trace], layout=layout)
iplot(fig)
```
