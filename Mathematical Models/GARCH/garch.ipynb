# We will use GARCH models to forecast volatilities of various stocks and stock indices.

## Stocks

### Apple

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from arch import arch_model

# Fetching the stock data
ticker = "AAPL"
data = yf.download(ticker, start="2020-01-01", end="2023-01-01")

# Preprocessing the data
data['Date'] = data.index
data.reset_index(drop=True, inplace=True)
data = data[['Date', 'Close']]

# Calculate daily returns
data['Return'] = data['Close'].pct_change().dropna() * 100

# Drop NaN values resulting from the percentage change calculation
data = data.dropna()

# Fit a GARCH(1,1) model
returns = data['Return']
model = arch_model(returns, vol='Garch', p=1, q=1)
model_fit = model.fit(disp='off')

# Forecasting
forecast_horizon = 5
forecast = model_fit.forecast(horizon=forecast_horizon)
mean_forecast = forecast.mean['h.1'].iloc[-1]
variance_forecast = forecast.variance['h.1'].iloc[-1]
print(f"Mean forecast: {mean_forecast}")
print(f"Variance forecast: {variance_forecast}")

# Plotting the stock price
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Close'], label='Close')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title(f'{ticker} Closing Price')
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(data['Date'], data['Return'], label='Returns')
plt.xlabel('Date')
plt.ylabel('Daily Returns (%)')
plt.title(f'{ticker} Daily Returns')
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()

# Plot forecasted volatility
plt.figure(figsize=(10, 6))
plt.plot(forecast.variance.iloc[-1], label='Forecasted Variance')
plt.xlabel('Horizon')
plt.ylabel('Forecasted Variance')
plt.title(f'Forecasted Variance over the Horizon for {ticker}')
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
```
