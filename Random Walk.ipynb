## Random Walk
<br>
A random walk is a stochastic process that describes a path of random steps on a mathematical space. It's also known as a drunkard's walk.
<br><br>
The random walk hypothesis is a financial theory stating that stock market prices evolve according to a random walk and thus cannot be predicted.


### Random Walk in 1D

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def randomwalk1D(n):
    x, y = 0, 0
    
    timepoints = np.arange(n + 1)
    positions = [y]

    directions = ["up", "down"]
    for i in range(1, n + 1):

        step = random.choice(directions)
                
        if step == "up":
            y += 1
        elif step == "down":
            y -= 1
        
        positions.append(y)

    return timepoints, positions
```

```python
num = 1000
rw1 = randomwalk1D(num)
rw2 = randomwalk1D(num)
rw3 = randomwalk1D(num)
```

```python
plt.plot(rw1[0], rw1[1], 'r-', label="rw1")
plt.plot(rw2[0], rw2[1], 'g-', label="rw2")
plt.plot(rw3[0], rw3[1], 'b-', label="rw3")
plt.title('Random Walk in 1D')

plt.show()
```

### Random Walk in 2D

```python
def randomwalk2D(n):
    
    x = np.zeros(n)
    y = np.zeros(n)
    directions = ["up", "down", "left", "right"]
    for i in range(1, n):
        
        step = random.choice(directions)
        
        if step == "right":
            x[i] = x[i - 1] + 1
            y[i] = y[i - 1]
        elif step == "left":
            x[i] = x[i - 1] - 1
            y[i] = y[i - 1]
        elif step == "up":
            x[i] = x[i - 1]
            y[i] = y[i - 1] + 1
        elif step == "down":
            x[i] = x[i - 1]
            y[i] = y[i - 1] - 1
    
    return x, y
```

```python
x_data, y_data = randomwalk2D(num)

plt.title("Random Walk in 2D")
plt.plot(x_data, y_data, color='green')
plt.show()
```

### Random Walk in 3D

```python
def randomwalk3D(n):
    x, y, z = np.zeros(n), np.zeros(n), np.zeros(n)
    directions = ["up", "down", "left", "right", "in", "out"]
    for i in range(1, n):
        step = random.choice(directions)
        if step == "right":
            x[i] = x[i - 1] + 1
            y[i] = y[i - 1]
            z[i] = z[i - 1]
        elif step == "left":
            x[i] = x[i - 1] - 1
            y[i] = y[i - 1]
            z[i] = z[i - 1]
        elif step == "up":
            x[i] = x[i - 1]
            y[i] = y[i - 1] + 1
            z[i] = z[i - 1]
        elif step == "down":
            x[i] = x[i - 1]
            y[i] = y[i - 1] - 1
            z[i] = z[i - 1]
        elif step == "in":
            x[i] = x[i - 1]
            y[i] = y[i - 1]
            z[i] = z[i - 1] - 1
        elif step == "out":
            x[i] = x[i - 1]
            y[i] = y[i - 1]
            z[i] = z[i - 1] + 1
    return x, y, z
```

```python
x_data, y_data, z_data = randomwalk3D(num)

ax = plt.subplot(1, 1, 1, projection='3d')
ax.plot(x_data, y_data, z_data, alpha=0.9)
ax.scatter(x_data[-1], y_data[-1], z_data[-1])

plt.title('Random Walk in 3D')

plt.show()
```
