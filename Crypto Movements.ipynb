### Returns on NIFTY 50

```python

nse_daily_returns = nse['Adj Close'].pct_change().dropna()
nse_monthly_returns = nse['Adj Close'].resample('M').ffill().pct_change().dropna()
```

```python
nse_daily_returns.head()
```

```python
nse_monthly_returns.head()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.plot(nse_daily_returns)
ax1.set_xlabel("Date")
ax1.set_ylabel("Percent")
ax1.set_title("NIFTY 50 Daily Return")

plt.xticks(rotation=45)
fig.tight_layout()
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.plot(nse_monthly_returns)
ax1.set_xlabel("Date")
ax1.set_ylabel("Percent")
ax1.set_title("NIFTY 50 Monthly Return")

plt.xticks(rotation=45)
fig.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(8, 5))
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
nse_daily_returns.plot.hist(bins = 60)
ax1.set_xlabel("Daily returns %")
ax1.set_ylabel("Percent")
ax1.set_title("NIFTY 50 Daily Return")
ax1.text(-0.12,20,"Extreme Low\nReturns")
ax1.text(0.06,20,"Extreme High\nReturns")

plt.xticks(rotation=45)
plt.show()
```

#### Let's calculate a volatility function.

```python
def get_daily_volatility(df):

    # Calculate daily returns
    daily_returns = df['Adj Close'].pct_change().dropna()

    # Calculate daily volatility (standard deviation of daily returns)
    daily_volatility = daily_returns.std()

    return daily_volatility

def get_monthly_volatility(df):

    # Calculate monthly returns
    monthly_returns = df['Adj Close'].resample('M').ffill().pct_change().dropna()

    # Calculate monthly volatility (standard deviation of monthly returns)
    monthly_volatility = monthly_returns.std()

    return monthly_volatility
```

```python
x = get_daily_volatility(ni), get_daily_volatility(sp500), get_daily_volatility(nasdaq), get_daily_volatility(nse)
x
```

```python
y = get_daily_volatility(btc), get_daily_volatility(eth), get_daily_volatility(bnb)
y
```

```python
z = np.append(np.array(x), np.array(y))
z
```

### Plotting

```python
plt.scatter(['NIKKEI','S&P500','Nasdaq','NIFTY 50','Bitcoin','Ethereum','BNB'],z,s=10,alpha=0.75,c='g')
plt.ylabel('Volatiliy')
```

#### Thus, cryptos have a higher volatility than stocks.
