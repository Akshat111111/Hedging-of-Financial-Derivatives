**Step 2: Data Preprocessing**


**For LSTM, we need to scale the data. For SVR and KNN, we also need to preprocess the data accordingly.**

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
# Convert the Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Set the Date column as the index
df.set_index('Date', inplace=True)

# Sort the data by date
df = df.sort_index()

# Select the 'Close' column for prediction
data = df[['Close']]
```

```python
# Scale the data for LSTM
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(data)
```

```python
# Prepare the data for LSTM
def create_dataset(dataset, time_step=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - time_step - 1):
        a = dataset[i:(i + time_step), 0]
        dataX.append(a)
        dataY.append(dataset[i + time_step, 0])
    return np.array(dataX), np.array(dataY)

time_step = 60
X, y = create_dataset(scaled_data, time_step)
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
# Reshape input for LSTM [samples, time steps, features]
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

**Step 3: Implementing the Models**

**LSTM**

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
```

```python
# Build the LSTM model
model_lstm = Sequential()
model_lstm.add(LSTM(50, return_sequences=True, input_shape=(time_step, 1)))
model_lstm.add(LSTM(50, return_sequences=False))
model_lstm.add(Dense(25))
model_lstm.add(Dense(1))
```

```python
# Compile the model
model_lstm.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the model
history = model_lstm.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Predicting the closing price using the LSTM model
train_predict = model_lstm.predict(X_train)
test_predict = model_lstm.predict(X_test)
```

```python
# Inverse transform to get actual prices
train_predict = scaler.inverse_transform(train_predict)
test_predict = scaler.inverse_transform(test_predict)
y_train = scaler.inverse_transform(y_train.reshape(-1, 1))
y_test = scaler.inverse_transform(y_test.reshape(-1, 1))
```

```python
# Plotting the results
plt.figure(figsize=(12, 6))
plt.plot(df.index, data, label='Actual Price')
plt.plot(df.index[:len(train_predict)], train_predict, label='Train Predict')
plt.plot(df.index[len(train_predict) + time_step + 1:], test_predict, label='Test Predict')
plt.title('LSTM Model')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

**SVR Model**

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
```

```python
# Prepare the data for SVR (we don't scale it here as SVR has its own scaling)
X_svr = df[['Open', 'High', 'Low', 'Close', 'Volume']].values
y_svr = df['Close'].values
```

```python
# Split the data into training and testing sets
X_train_svr, X_test_svr, y_train_svr, y_test_svr = train_test_split(X_svr, y_svr, test_size=0.2, shuffle=False)
```

```python
# Build and train the SVR model
model_svr = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=0.1)
model_svr.fit(X_train_svr, y_train_svr)

# Predicting the closing price using the SVR model
train_predict_svr = model_svr.predict(X_train_svr)
test_predict_svr = model_svr.predict(X_test_svr)
```

```python
# Evaluating the model
mse_svr = mean_squared_error(y_test_svr, test_predict_svr)
print(f'SVR Mean Squared Error: {mse_svr}')
```

```python
# Plotting the results
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Close'], label='Actual Price')  # Use df['Close'] for actual prices
plt.plot(df.index[:len(train_predict_svr)], train_predict_svr, label='Train Predict')

# Adjust the index for the test predictions to match the length of test_predict_svr
test_index = df.index[len(train_predict_svr): len(train_predict_svr) + len(test_predict_svr)]
plt.plot(test_index, test_predict_svr, label='Test Predict')

plt.title('SVR Model')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

**KNN Model**

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
# Prepare the data for KNN (we don't scale it here as KNN has its own scaling)
X_knn = df[['Open', 'High', 'Low', 'Close', 'Volume']].values
y_knn = df['Close'].values
```

```python
# Split the data into training and testing sets
X_train_knn, X_test_knn, y_train_knn, y_test_knn = train_test_split(X_knn, y_knn, test_size=0.2, shuffle=False)
```

```python
# Build and train the KNN model
model_knn = KNeighborsRegressor(n_neighbors=5)
model_knn.fit(X_train_knn, y_train_knn)

# Predicting the closing price using the KNN model
train_predict_knn = model_knn.predict(X_train_knn)
test_predict_knn = model_knn.predict(X_test_knn)
```

```python
# Evaluating the model
mse_knn = mean_squared_error(y_test_knn, test_predict_knn)
print(f'KNN Mean Squared Error: {mse_knn}')
```

```python
# Plotting the results
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Close'], label='Actual Price')  # Use df['Close'] for actual prices
plt.plot(df.index[:len(train_predict_knn)], train_predict_knn, label='Train Predict')

# Adjust the index for the test predictions to align with the actual test data
test_index = df.index[len(train_predict_knn):]
plt.plot(test_index, test_predict_knn, label='Test Predict')

plt.title('KNN Model')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```
