**Step 3: Data Preprocessing, split the dataset into training and testing sets**

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
# Convert 'Date' to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Sort the data by date
df = df.sort_values('Date')
```

```python
# Drop any rows with missing values
df = df.dropna()

# Feature selection
features = df[['Open', 'High', 'Low', 'Close', 'Volume']]
```

```python
# Scaling features
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_features = scaler.fit_transform(features)
```

```python
# Splitting data into training and testing sets
train_size = int(len(scaled_features) * 0.8)
train, test = scaled_features[:train_size], scaled_features[train_size:]
```

```python
# Separating features and target
X_train, y_train = train[:, :-1], train[:, -1]
X_test, y_test = test[:, :-1], test[:, -1]
```

**Step 4: Comparing and Predicting with differnt Models**

**LSTM Model**

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
```

```python
# Reshape the data for LSTM model
X_train_lstm = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test_lstm = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
```

```python
# LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train_lstm.shape[1], 1)))
model.add(LSTM(units=50))
model.add(Dense(1))
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the model
model.fit(X_train_lstm, y_train, epochs=50, batch_size=32)
```

```python
# Assuming your scaler was fitted with 5 features and you only care about the first one
predicted_stock_price = model.predict(X_test_lstm)

# Create an array of the correct shape with only the predicted stock price and zeros for other features
predicted_stock_price_full = np.zeros((predicted_stock_price.shape[0], 5))
predicted_stock_price_full[:, 0] = predicted_stock_price[:, 0]

# Inverse transform
predicted_stock_price_full = scaler.inverse_transform(predicted_stock_price_full)

# Extract only the first column (predicted stock prices)
predicted_stock_price = predicted_stock_price_full[:, 0]
```

```python
# Visualizing the results
plt.figure(figsize=(10, 6))
plt.plot(df['Date'][train_size:], df['Close'][train_size:], color='blue', label='Actual Hero MotoCorp Stock Price')
plt.plot(df['Date'][train_size:], predicted_stock_price, color='red', label='Predicted Hero MotoCorp Stock Price')
plt.title('Hero MotoCorp Stock Price Prediction using LSTM')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```

**SVM Model**


```python
from sklearn.svm import SVR
```

```python
# SVM model
svm_model = SVR(kernel='rbf')
svm_model.fit(X_train, y_train)
```

```python
# Predict the stock price using SVM
svm_predicted = svm_model.predict(X_test)

# Create an array with the correct shape (492, 5) and populate the first column with predictions
svm_predicted_full = np.zeros((svm_predicted.shape[0], 5))
svm_predicted_full[:, 0] = svm_predicted

# Inverse transform the full array
svm_predicted_full = scaler.inverse_transform(svm_predicted_full)

# Extract only the first column which corresponds to the stock price
svm_predicted_stock_price = svm_predicted_full[:, 0]
```

```python
# Visualizing the results
plt.figure(figsize=(10, 6))
plt.plot(df['Date'][train_size:], df['Close'][train_size:], color='blue', label='Actual Hero MotoCorp Stock Price')
plt.plot(df['Date'][train_size:], svm_predicted, color='red', label='Predicted Hero MotoCorp Stock Price')
plt.title('Hero MotoCorp Stock Price Prediction using SVM')
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```
